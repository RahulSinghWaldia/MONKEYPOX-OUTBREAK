```python

import numpy as np # linear algebra
import pandas as pd # data processing
```

```python
df=pd.read_csv(r"/content/sample_data/Monkeypox.csv")
```

```python
df.head(10)
```

```python
import missingno as msno
```

```python
msno.matrix(df)
```

```python
for i in df.columns:
    if (((df[i].isnull().sum())/(df[i].shape)[0])*100)>70:
        df=df.drop([i],axis=1)
```

```python
df.head(5)
```

```python
df.groupby(df['Country']).count().sort_values(by='Status',ascending=False)['Status'].plot(kind='bar',color='red',figsize=(18,10))
```

```python
value_list=df.groupby(df['Country']).count().sort_values(by='Status',ascending=False)['Status']
data_value=pd.DataFrame(value_list.values,columns=['confirmed(rank)'])
data_column=pd.DataFrame(value_list.keys(),columns=['Country'])
new_data_frm=pd.concat([data_column,data_value],axis=1)
new_data_frm
```
